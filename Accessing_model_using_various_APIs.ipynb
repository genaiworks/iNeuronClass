```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
tokenizer = AutoTokenizer.from_pretrained("google/gemma-7b")
```

```python
input_text = "can you tell me about the capital of india?."
input_ids = tokenizer(input_text, return_tensors="pt").to("cuda")
```

```python
outputs = model.generate(**input_ids)
```

# Google API

```python
from google.colab import userdata
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
```

```python
import google.generativeai as genai
```

```python
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
for model in genai.list_models():
  print(model)
```

```python
for model in genai.list_models():
  if "generateContent" in model.supported_generation_methods:
    print(model.name)
```

```python
def load_model(model_name):
  if model_name=="gemini-pro":
    #text to text
    llm=genai.GenerativeModel("gemini-pro")
  else:
    #image to text
    llm=genai.GenerativeModel("gemini-pro-vision")
  return llm
```

```python
model_text=load_model("gemini-pro")
```

```python
model_vision=load_model("gemini-pro-vision")
```

```python
response=model_text.generate_content("what is a capital of chaina give me a detail about it")
```

```python
print(response.text)
```
